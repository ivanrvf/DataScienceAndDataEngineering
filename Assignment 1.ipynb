Below is the code for web scrapper. I am using this "https://www.trivago.com" website for web scraping.
1. List the links is external or internal 
2. List the http status of these links
3. Saving in excel file

```python
from __future__ import print_function
import urllib3 #  pip install -i https://pypi.anaconda.org/pypi/simple urllib3
import unicodedata
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
import re
import datetime
warnings.filterwarnings('ignore')
url='https://www.trivago.com'
regURL = re.search('[a-z]+\.[a-z]+\.[a-z]+', url)
http = urllib3.PoolManager()
response = http.request('GET',url)
```

Code to check if links are external or internal and valid

```python
soup=BeautifulSoup(response.data, "html.parser")
linkType = "";
status = "";
linkURLDict = {}
i = 0
for link in soup.find_all('a'):
    hrefURL = link.get('href')
    text = link.contents
    try:
        response = http.request('GET',hrefURL)
        status = response.status
        if status == 200:
            status = True
            hrefWebsiteURL = re.search('[a-z]+\.[a-z]+\.[a-z]+', hrefURL)
            if hrefWebsiteURL.group(0) == regURL.group(0):
                linkType = "Internal"
            else:
                linkType = "External"
        else:
            status = False
            linkType = "Cannot identify"
    except:
        status = False
        linkType = "Exception"
    listDict = {}
    listDict.update({"domainType":linkType,"httpRequestStatus":status,"URL":hrefURL,"Associated Text":text,"Timestamp":datetime.datetime.now()})
    linkURLDict.update({i:listDict})
    i+=1
LinkURLDF = pd.DataFrame.from_dict(linkURLDict,orient = 'index')
LinkURLDF
writer = pd.ExcelWriter('WebScraping.xlsx')
LinkURLDF.to_excel(writer,sheet_name='Sheet1',index=True)
writer.save()
```
